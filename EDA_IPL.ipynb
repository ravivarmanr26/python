```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example DataFrame with categorical variables
data = pd.DataFrame({
    'Category1': ['A', 'B', 'A', 'C', 'B'],
    'Category2': ['X', 'Y', 'Y', 'X', 'Z'],
    'Numeric1': [10, 20, 15, 25, 30],
    'Numeric2': [5, 15, 10, 20, 25]
})

# One-hot encode categorical variables
data_encoded = pd.get_dummies(data[['Category1', 'Category2']])
print(data_encoded)

# Concatenate the numerical and encoded categorical columns
data_combined = pd.concat([data[['Numeric1', 'Numeric2']], data_encoded], axis=1)

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data_combined.corr(), annot=True, cmap='viridis', fmt=".2f")
plt.show()
```
